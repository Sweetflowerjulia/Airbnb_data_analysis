# Airbnb Review Sentiment

## NLP: Sentiment analysis on reviews
* Conduct sentiment classification for review comments and identify the review is positive, negative or neutral.
* Use textblob package to get sentiment of the text.

```python
import pandas as pd
import numpy as np
reviews = pd.read_csv("Berlin Airbnb Dataset/reviews.csv")
```

```python
from textblob import TextBlob
import re 

def get_sentiment(text):
    text = re.sub('[^\w\s]', '', text) # clean text
    analysis = TextBlob(text) # get sentiment
    
    if analysis.sentiment.polarity > 0: 
        return 'positive'
    elif analysis.sentiment.polarity == 0: 
        return 'neutral'
    else: 
        return 'negative'
```

```python
reviews['sentiment'] = reviews.comments.apply(lambda x: get_sentiment(str(x)))
```
